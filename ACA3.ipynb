```python
#selecciona campos que quiero mostrar, cual quiero sumar y por cual quiero agrupar
agregado = dataSept.groupby(['COD_PAI4','PAIS'])['FOBDOL'].sum().reset_index()
#renombro columna de suma
agregado = agregado.rename(columns={'FOBDOL':'FOBDOL'})
#ordeno de mayor a menor
agregado = agregado.sort_values(by='FOBDOL',ascending=False).head(5)
#elimino la columna de indice
agregado = agregado.reset_index(drop=True)
# Imprimir o trabajar con el DataFrame resultante
print(agregado)
```

```python
#crea matriz para el where
valores_a_buscar = ['USA', 'PAN', 'IND', 'MEX', 'BRA']
#trae de la tabla solo los campos para esos valores
resultados_filtrados = dataSept[dataSept['COD_PAI4'].isin(valores_a_buscar)]
# Imprimir o trabajar con el DataFrame resultante
print(resultados_filtrados)
```

```python
# Exportar el DataFrame resultante a un archivo CSV
resultados_filtrados.to_csv('resultados_filtrados.csv', index=False)
```
