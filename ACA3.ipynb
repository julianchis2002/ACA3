##Union de los dataframes para trabajar como uno unificado

## Creacion de dataframes para analisis varios

```python
#Totales en pesos por pais y mes
dataTotal0 = dataTotal.groupby(['FECH'])['FOBPES'].sum().reset_index()
dataTotal0['FECH'] = dataTotal0['FECH'].astype(str)
plt.figure()
plt.bar(dataTotal0['FECH'], dataTotal0['FOBPES'], color='skyblue')
plt.show()
```

```python
#Totales en pesos por pais y salida
dataTotal1 = dataTotal.loc[:,['COD_PAI4','COD_SAL','FOBPES']]
dataTotal1.pivot_table(index='COD_SAL',columns='COD_PAI4',values='FOBPES',aggfunc=sum,fill_value=0).T
```

```python
#Selecciona los 5 primeros paises en exportaciones totales (Sep - Oct)
dataTotal2 = dataTotal.groupby(['COD_PAI4','PAIS'])['FOBPES'].sum().reset_index()
dataTotal2 = dataTotal2.sort_values(by='FOBPES',ascending=False).head(5)
dataTotal2 = dataTotal2.reset_index(drop=True)
print(dataTotal2)
plt.figure()
plt.bar(dataTotal2['COD_PAI4'], dataTotal2['FOBPES'], color='skyblue')
plt.show()
```

```python
#Curtosis sobre la data total
dataTotal.kurtosis()
```

##Selecciona la data para los 5 paises

```python
top = ['USA', 'PAN', 'IND', 'ECU', 'MEX']
dataFiltrada = dataTotal[dataTotal['COD_PAI4'].isin(top)]
print(dataFiltrada)
```

```python
# Exportar el DataFrame resultante a un archivo CSV
dataFiltrada.to_csv('resultados_filtrados.csv', index=False)
```
